üìö Un mot important sur la m√©thode scientifique

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("temperature.csv") \
       .merge(pd.read_csv("co2.csv"), on="Ann√©e") \
       .merge(pd.read_csv("energie.csv"), on="Ann√©e") \
       .merge(pd.read_csv("accidents.csv"), on="Ann√©e") \
       .merge(pd.read_csv("penalties.csv"), on="Ann√©e") \
       .merge(pd.read_csv("salles_sport.csv"), on="Ann√©e") \
       .merge(pd.read_csv("films_action.csv"), on="Ann√©e")

# Visualisation
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
axs = axs.flatten()

variables = df.columns[1:]
colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'black']

for i, col in enumerate(variables):
    axs[i].plot(df['Ann√©e'], df[col], color=colors[i % len(colors)])
    axs[i].set_title(col.replace("_", " "))
    axs[i].set_xlabel("Ann√©e")

plt.tight_layout()
plt.show()

# Affichage de la matrice de corr√©lation
corr_matrix = df.drop(columns=["Ann√©e"]).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corr√©lation")
plt.show()
```
